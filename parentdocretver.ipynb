```python
from langchain.retrievers import ParentDocumentRetriever
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate,ChatPromptTemplate,HumanMessagePromptTemplate,SystemMessagePromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import TextLoader
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
import os
import chromadb
from chromadb.config import Settings
from dotenv import load_dotenv
from pprint import pprint 
```

```python
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")

llm = GoogleGenerativeAI(
    model = "gemini-1.5-flash",google_api_key=os.environ["GOOGLE_API_KEY"])


client = chromadb.HttpClient(host="127.0.0.1", settings=Settings(allow_reset=True))
db = Chroma(client= client, embedding_function=embeddings)


retriver = db.as_retriever()
```

```python


loaders = [
    PyPDFLoader("./pdf-docs/mml-book.pdf")
]
docs = []
for l in loaders:
    docs.extend(l.load())

print(f"Number of documents: {len(docs)}")
```

```python
childsplitter = RecursiveCharacterTextSplitter(chunk_size=400,chunk_overlap=50)





```

```python
print(len(docs[6].page_content))
```

```python
vectorestore = Chroma(collection_name="mml_book",
                      embedding_function = embeddings,
                      persist_directory = "./dbbook",
                      )
```

```python
store = InMemoryStore()

full_doc_retriever = ParentDocumentRetriever(
                                            vectorstore=vectorestore,
                                            docstore=store,
                                            child_splitter=childsplitter,
                                            )
```

```python
full_doc_retriever.add_documents(docs, ids=None)
```

```python
list(store.yield_keys())
```

```python
sub_docs = vectorestore.similarity_search("what is Linear Algebra", k=5)
pprint(sub_docs[4].page_content)
```

```python
retrieved_docs = full_doc_retriever.get_relevant_documents("what is  Linear Algebra")
print(retrieved_docs[1].page_content)
```

```python
child_chunk = sub_docs[0].page_content
parent_doc = retrieved_docs[0].page_content
```

```python
print("Length of child chunk: ", len(child_chunk))
print("Length of parent doc: ", len(parent_doc))
```

```python
from langchain.chains import RetrievalQA


qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=full_doc_retriever)
```

```python
response = qa.invoke("What is an Generative AI?")
response
```
